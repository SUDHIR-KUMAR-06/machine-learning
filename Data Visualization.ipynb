```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4])
y=x**2+5
print(x,y)
```

# plt.style.available
- it will show all available types of plotchart



```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y)
plt.title('my line plot')

plt.show()
```

```python
plt.plot(x,y,label="stock price")
plt.plot(x,x**2,label="stock price1")
plt.title("my plot chart")
plt.xlabel("timestamp")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label="stock price")
plt.scatter(x,x**2,label="stock price1")
plt.title("my plot chart")
plt.xlabel("timestamp")
plt.ylabel("price")
plt.legend()
plt.show()
```
